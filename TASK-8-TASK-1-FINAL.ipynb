```python
def trans_german(str1):
    c=str1.split()
    trans_ls=[]
    for i in range (len(c)):
        trans_ls.append(to_german.translate(c[i]))
        res_g=' '.join(trans_ls)
    return res_g
```

```python
def trans_spanish(str1):
    c=str1.split()
    trans_ls=[]
    for i in range (len(c)):
        trans_ls.append(to_Spanish.translate(c[i]))
        res_g=' '.join(trans_ls)
    return res_g
```

```python
def trans_french(str1):
    c=str1.split()
    trans_ls=[]
    for i in range (len(c)):
        trans_ls.append(to_french.translate(c[i]))
        res_g=' '.join(trans_ls)
    return res_g
```
